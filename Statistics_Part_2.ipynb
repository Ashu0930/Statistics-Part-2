1.  What is hypothesis testing in statistics?

-> Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It helps determine whether there is enough evidence to support a specific belief or claim about a population.

#2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

->
Null Hypothesis (H0):

  - States that there is no relationship, no difference, or no effect between variables or groups being tested.
  - It's the default assumption that researchers aim to disprove or reject.

  Examples :
  - There is no difference in the average height between men and women.
  - The new drug has no effect on blood pressure.
  - There is no correlation between hours of study and exam scores

Alternative Hypothesis (H1 or Ha):

  - States that there is a relationship, a difference, or an effect between variables or groups being tested.
  - It's the hypothesis that researchers are trying to prove or support.

  Examples :
  - There is a difference in the average height between men and women.
  - The new drug has an effect on blood pressure.
  - There is a correlation between hours of study and exam scores.   

Key Differences:

  - Purpose: The null hypothesis is the default assumption, while the alternative hypothesis is the researcher's claim.
  - Testing: Researchers use statistical tests to determine whether there's enough evidence to reject the null hypothesis in favor of the alternative hypothesis.
  - Notation: The null hypothesis is often denoted as H0, and the alternative hypothesis as H1 or Ha  

#3. What is the significance level in hypothesis testing, and why is it important?

-> In hypothesis testing, the significance level, often denoted as alpha (α), is the threshold for determining whether results are statistically significant. It's the probability of rejecting the null hypothesis when it's actually true, also known as a Type I error. A common significance level is 0.05 (5%), meaning there's a 5% chance of concluding a difference exists when it doesn't.

Why is it important :

1. Sets the bar for evidence:

  - It defines how much evidence you need to reject H₀.

  - If p-value < α, you reject the null hypothesis.

2. Controls error risk:

  - Helps manage the risk of false positives (thinking there's an effect when there isn't one).

3. Affects study conclusions:

  - A smaller α means you need stronger evidence to claim a real effect, reducing the chance of being wrong but increasing the chance of missing real effects (Type II errors).



#4. What does a P-value represent in hypothesis testing?

-> In hypothesis testing, a P-value represents the probability of observing results as extreme as, or more extreme than, the ones obtained from a statistical test, assuming the null hypothesis is true. Essentially, it measures the strength of evidence against the null hypothesis. A smaller P-value indicates stronger evidence against the null hypothesis, suggesting that the observed results are unlikely to have occurred by chance.

#5. How do you interpret the P-value in hypothesis testing?

-> The p-value indicates the probability of observing results as extreme as, or more extreme than, the ones obtained, assuming the null hypothesis is true. A small p-value (typically less than 0.05) suggests strong evidence against the null hypothesis, leading to its rejection in favor of the alternative hypothesis.

#6. What are Type 1 and Type 2 errors in hypothesis testing?

-> A Type I error (false positive) occurs when the null hypothesis is incorrectly rejected, and a Type II error (false negative) occurs when the null hypothesis is incorrectly failed to be rejected. Put simply, Type I errors lead to concluding there's a difference when there isn't, while Type II errors lead to missing a difference that actually exists.

#7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

-> In hypothesis testing, a one-tailed test, also known as a directional test, is used when the alternative hypothesis specifies a direction for the difference between the null hypothesis and the sample results (e.g., "more than," "less than"). A two-tailed test, on the other hand, is used when the alternative hypothesis does not specify a direction and simply tests for a difference between the null hypothesis and the sample results in either direction.

1. Direction of the Alternative Hypothesis:
  - One-tailed: Specifies a direction (e.g., > or <).
  - Two-tailed: Does not specify a direction (e.g., ≠).

2. Significance Level (α):

  - One-tailed: The entire significance level (α) is allocated to one tail of the distribution.
  - Two-tailed: The significance level (α) is split between both tails of the distribution.

3. Critical Regions:

  - One-tailed: Has one critical region (either in the left tail or right tail).
  - Two-tailed: Has two critical regions (one in each tail).

#10. What is the T-distribution, and when should it be used instead of the normal distribution?

-> The T-distribution (also called Student’s t-distribution) is a probability distribution used instead of the normal distribution when:

 1. The sample size is small (typically n < 30), and
 2. The population standard deviation (σ) is unknown — you only have the sample standard deviation (s).

Why Use the T-distribution?

When σ is unknown and n is small, the sample mean has more variability. The T-distribution accounts for this extra uncertainty by having "fatter tails" (more spread out) than the standard normal (Z) distribution.

As sample size increases, the T-distribution gets closer to the normal distribution.

#13. What is the relationship between Z-test and T-test in hypothesis testing?

-> While both the Z-test and T-test are used to test hypotheses about population means, they are used in different situations based on the known vs. unknown population parameters and sample size.

Here’s how they relate:

1. Underlying Assumptions:

a)  Z-test:

  - Population standard deviation (σ) is known.
  - Large sample size (n ≥ 30), or the population distribution is normal.

b) T-test:

  - Population standard deviation (σ) is unknown — you only have the sample standard deviation (s).
  - Small sample size (n < 30), and the population distribution is approximately normal.
  

2. Distribution Used:

a) Z-test uses the Z-distribution (standard normal distribution).

  - It has a fixed shape with mean = 0 and standard deviation = 1.

b) T-test uses the T-distribution.

  - The T-distribution is similar to the normal distribution, but has fatter tails, which accounts for the extra uncertainty when estimating the population standard deviation from a small sample.
  - As the sample size increases, the T-distribution approaches the Z-distribution (the larger the sample, the less the uncertainty about the population standard deviation).

  - A smaller sample size.
  - More variability in the data.

- The narrower the interval, the more precise the estimate is, typically resulting from:

  - A larger sample size.
  - Less variability in the data.

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
import numpy as np
from scipy import stats

# Function to perform Z-test
def z_test(sample_data, population_mean, population_std, sample_size):
    # Calculate the sample mean
    sample_mean = np.mean(sample_data)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate the p-value
    p_value = stats.norm.sf(abs(z_statistic)) * 2  # Two-tailed test

    # Return results
    return z_statistic, p_value

# Sample data (e.g., sample of test scores or measurements)
sample_data = [80, 82, 78, 85, 88, 90, 79, 84, 86, 81]

# Known population parameters
population_mean = 85  # Known population mean
population_std = 5    # Known population standard deviation
sample_size = len(sample_data)

# Perform the Z-test
z_statistic, p_value = z_test(sample_data, population_mean, population_std, sample_size)

# Print the results
print(f"Z-statistic: {z_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

#4.  Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate random data (e.g., test scores or measurements)
np.random.seed(42)  # For reproducibility
sample_size = 100  # Sample size
population_mean = 75  # Known population mean
population_std = 10   # Known population standard deviation

# Simulate a random sample from the population
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Step 2: Perform the Two-Tailed Z-test
def z_test(sample_data, population_mean, population_std, sample_size):
    # Calculate the sample mean
    sample_mean = np.mean(sample_data)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate the p-value for a two-tailed test
    p_value = stats.norm.sf(abs(z_statistic)) * 2  # Multiply by 2 for two-tailed test

    return z_statistic, p_value

# Perform the Z-test
z_statistic, p_value = z_test(sample_data, population_mean, population_std, sample_size)

# Step 3: Print the results
print(f"Sample Mean: {np.mean(sample_data):.2f}")
print(f"Z-statistic: {z_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Visualize the decision region and the Z-distribution
alpha = 0.05  # Significance level for two-tailed test

# Critical values for Z-test at alpha = 0.05 (two-tailed)
z_critical = stats.norm.ppf(1 - alpha / 2)

# Generate values for plotting the Z-distribution
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

# Plot the Z-distribution
plt.plot(x, y, label='Z-distribution', color='blue')

# Fill the critical regions (rejection regions)
plt.fill_between(x, y, 0, where=(x <= -z_critical), color='red', alpha=0.3, label='Rejection Region (Left)')
plt.fill_between(x, y, 0, where=(x >= z_critical), color='red', alpha=0.3, label='Rejection Region (Right)')

# Mark the Z-statistic on the plot
plt.axvline(z_statistic, color='green', linestyle='--', label=f'Z-statistic: {z_statistic:.2f}')
plt.axvline(-z_statistic, color='green', linestyle='--')

# Labeling the plot
plt.title('Two-Tailed Z-test: Decision Region')
plt.xlabel('Z')
plt.ylabel('Density')
plt.legend()

# Display the plot
plt.show()

# Step 5: Interpret the results
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Function to simulate and calculate Type I and Type II errors
def visualize_errors(population_mean, population_std, sample_size, true_mean, alpha=0.05):
    # Step 1: Simulate the null hypothesis (H0) and alternative hypothesis (H1) distributions
    # H0: The population mean is as expected
    # H1: The true population mean is different from H0

    # Null hypothesis distribution (assuming population mean is population_mean)
    null_dist = np.random.normal(population_mean, population_std, size=100000)

    # Alternative hypothesis distribution (assuming true mean is true_mean)
    alt_dist = np.random.normal(true_mean, population_std, size=100000)

    # Step 2: Z-test under H0 (assuming population mean is population_mean)
    z_critical = stats.norm.ppf(1 - alpha / 2)  # Critical value for two-tailed test
    z_statistic = (true_mean - population_mean) / (population_std / np.sqrt(sample_size))  # Z-statistic under H1

    # Step 3: Calculate the Type I and Type II errors
    # Type I error: The probability of rejecting the null hypothesis when it is true
    type_1_error = 1 - stats.norm.cdf(z_critical) + stats.norm.cdf(-z_critical)

    # Type II error: The probability of failing to reject the null hypothesis when the alternative is true
    type_2_error = stats.norm.cdf(z_critical - z_statistic) - stats.norm.cdf(-z_critical - z_statistic)

    print(f"Type I Error (α): {type_1_error:.4f}")
    print(f"Type II Error (β): {type_2_error:.4f}")

    # Step 4: Visualization
    plt.figure(figsize=(10, 6))

    # Plotting the null hypothesis distribution
    x = np.linspace(population_mean - 4 * population_std, population_mean + 4 * population_std, 1000)
    y_null = stats.norm.pdf(x, population_mean, population_std)
    plt.plot(x, y_null, label='Null Hypothesis (H₀)', color='blue')

    # Plotting the alternative hypothesis distribution
    y_alt = stats.norm.pdf(x, true_mean, population_std)
    plt.plot(x, y_alt, label='Alternative Hypothesis (H₁)', color='green')

    # Highlight Type I Error region (rejection region under H₀)
    plt.fill_between(x, y_null, 0, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3, label='Type I Error Region')

    # Highlight Type II Error region (failure to reject under H₁)
    plt.fill_between(x, y_alt, 0, where=(x >= -z_critical + z_statistic) & (x <= z_critical + z_statistic), color='yellow', alpha=0.3, label='Type II Error Region')

    # Plot the critical z values (decision thresholds)
    plt.axvline(-z_critical + z_statistic, color='red', linestyle='--', label=f'-Z Critical + Z Statistic')
    plt.axvline(z_critical + z_statistic, color='red', linestyle='--', label=f'Z Critical + Z Statistic')

    # Labels and titles
    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend(loc='upper right')

    # Show the plot
    plt.show()

# Parameters
population_mean = 75  # Known population mean
population_std = 10   # Known population standard deviation
sample_size = 100     # Sample size
true_mean = 80        # True mean for the alternative hypothesis

# Call the function to visualize Type I and Type II errors
visualize_errors(population_mean, population_std, sample_size, true_mean)
```

#6.  Write a Python program to perform an independent T-test and interpret the results@

```python
import numpy as np
from scipy import stats

# Step 1: Simulate two independent samples (e.g., test scores from two different classes)
np.random.seed(42)  # For reproducibility
sample_size = 50    # Sample size for each group
mean_group_1 = 75   # Mean of group 1
mean_group_2 = 80   # Mean of group 2
std_dev = 10        # Standard deviation for both groups

# Generate random samples for two independent groups
group_1 = np.random.normal(loc=mean_group_1, scale=std_dev, size=sample_size)
group_2 = np.random.normal(loc=mean_group_2, scale=std_dev, size=sample_size)

# Step 2: Perform the Independent T-test
t_statistic, p_value = stats.ttest_ind(group_1, group_2)

# Step 3: Print the results
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpret the results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis: The means of the two groups are significantly different.")
else:
    print("Fail to reject the null hypothesis: The means of the two groups are not significantly different.")
```

#7.  Perform a paired sample T-test using Python and visualize the comparison results.

#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(sample_data, confidence_level=0.95):
    """
    Function to calculate the confidence interval for a sample mean.

    Parameters:
    - sample_data: Array of sample data.
    - confidence_level: Confidence level for the interval (default is 0.95 for 95% confidence).

    Returns:
    - (lower_bound, upper_bound): The lower and upper bounds of the confidence interval.
    """
    # Calculate sample mean and sample standard deviation
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation (ddof=1 for sample)
    sample_size = len(sample_data)

    # Get the Z-value for the desired confidence level (for a 95% CI, Z = 1.96)
    z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)

    # Calculate the margin of error
    margin_of_error = z_value * (sample_std / np.sqrt(sample_size))

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Sample data (mean=50, std=10)

# Calculate the confidence interval for the sample mean
confidence_interval = calculate_confidence_interval(sample_data, confidence_level=0.95)

print(f"Confidence Interval: {confidence_interval}")
```

#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayesian_inference(prior_h1, prior_h2, likelihood_e_given_h1, likelihood_e_given_h2):
    """
    Perform Bayesian inference to calculate the posterior probability using Bayes' Theorem.

    Parameters:
    - prior_h1: Prior probability of hypothesis H1 (e.g., having the disease).
    - prior_h2: Prior probability of hypothesis H2 (e.g., not having the disease).
    - likelihood_e_given_h1: Likelihood of evidence E given H1.
    - likelihood_e_given_h2: Likelihood of evidence E given H2.

    Returns:
    - posterior_h1: Posterior probability of hypothesis H1 (e.g., having the disease) given evidence E.
    - posterior_h2: Posterior probability of hypothesis H2 (e.g., not having the disease) given evidence E.
    """
    # Calculate the marginal likelihood (evidence probability)
    p_e = (likelihood_e_given_h1 * prior_h1) + (likelihood_e_given_h2 * prior_h2)

    # Apply Bayes' Theorem to compute the posterior probabilities
    posterior_h1 = (likelihood_e_given_h1 * prior_h1) / p_e
    posterior_h2 = (likelihood_e_given_h2 * prior_h2) / p_e

    return posterior_h1, posterior_h2


# Given data
prior_h1 = 0.01  # Prior probability of having the disease
prior_h2 = 0.99  # Prior probability of not having the disease
likelihood_e_given_h1 = 0.95  # Likelihood of test positive if patient has the disease
likelihood_e_given_h2 = 0.05  # Likelihood of test positive if patient doesn't have the disease

# Perform Bayesian Inference
posterior_h1, posterior_h2 = bayesian_inference(prior_h1, prior_h2, likelihood_e_given_h1, likelihood_e_given_h2)

# Output the results
print(f"Posterior Probability of Having Disease (H1 | E): {posterior_h1:.4f}")
print(f"Posterior Probability of Not Having Disease (H2 | E): {posterior_h2:.4f}")
```

#13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

```python
import numpy as np

def calculate_expected_frequencies(observed_data):
    """
    Calculate the expected frequencies for a Chi-square test based on observed data.

    Parameters:
    - observed_data: 2D numpy array (contingency table)

    Returns:
    - expected: 2D numpy array of expected frequencies
    """
    # Total number of observations (sum of all entries in the table)
    total = np.sum(observed_data)

    # Sum of each row (marginal sum)
    row_totals = np.sum(observed_data, axis=1, keepdims=True)

    # Sum of each column (marginal sum)
    col_totals = np.sum(observed_data, axis=0, keepdims=True)

    # Calculate expected frequencies
    expected = (row_totals * col_totals) / total

    return expected

# Example observed data (contingency table)
# Rows: Male, Female; Columns: Smoker, Non-smoker
observed_data = np.array([[50, 150],  # Male: Smoker, Non-smoker
                          [30, 170]])  # Female: Smoker, Non-smoker

# Calculate the expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

# Output the results
print("Observed Frequencies:")
print(observed_data)

print("\nExpected Frequencies:")
print(expected_frequencies)
```

#14.  Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
import numpy as np
import scipy.stats as stats

def goodness_of_fit_test(observed, expected):
    """
    Perform a goodness-of-fit test using the Chi-square statistic.

    Parameters:
    - observed: List or array of observed frequencies.
    - expected: List or array of expected frequencies.

    Returns:
    - chi2_stat: Chi-square statistic
    - p_value: P-value of the test
    """
    # Perform the Chi-square goodness-of-fit test
    chi2_stat, p_value = stats.chisquare(observed, expected)

    return chi2_stat, p_value

# Observed data from 60 rolls of a six-sided die
observed_data = np.array([10, 8, 12, 9, 11, 10])

# Expected data (assuming a fair die, each outcome should have equal probability)
expected_data = np.array([10, 10, 10, 10, 10, 10])

# Perform the goodness-of-fit test
chi2_stat, p_value = goodness_of_fit_test(observed_data, expected_data)

# Output the results
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not follow the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data follows the expected distribution.")
```

#16.  Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
import scipy.stats as stats

def f_test(sample1, sample2):
    """
    Perform an F-test to compare the variances of two samples.

    Parameters:
    - sample1: First sample data (array-like).
    - sample2: Second sample data (array-like).

    Returns:
    - F-statistic: The test statistic.
    - p-value: The p-value for the test.
    """
    # Calculate sample variances
    var1 = np.var(sample1, ddof=1)  # ddof=1 for sample variance (not population variance)
    var2 = np.var(sample2, ddof=1)

    # Calculate the F-statistic: ratio of the variances
    F_statistic = var1 / var2 if var1 >= var2 else var2 / var1  # Ensure F_statistic is >= 1

    # Degrees of freedom for the two samples
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1

    # Calculate the p-value from the F-distribution
    p_value = 2 * min(stats.f.cdf(F_statistic, df1, df2), 1 - stats.f.cdf(F_statistic, df1, df2))

    return F_statistic, p_value

# Example: Two random samples
np.random.seed(42)  # Set seed for reproducibility
sample1 = np.random.normal(0, 1, 100)  # Sample from a normal distribution with mean=0, std=1
sample2 = np.random.normal(0, 2, 100)  # Sample from a normal distribution with mean=0, std=2

# Perform the F-test
F_statistic, p_value = f_test(sample1, sample2)

# Output the results
print(f"F-statistic: {F_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")
```

#17. Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results.

```python
import numpy as np
import scipy.stats as stats

def anova_test(groups):
    """
    Perform a One-Way ANOVA test to compare the means of multiple groups.

    Parameters:
    - groups: A list of arrays, each representing a group of data.

    Returns:
    - F-statistic: The test statistic.
    - p-value: The p-value for the test.
    """
    # Perform the One-Way ANOVA
    F_statistic, p_value = stats.f_oneway(*groups)

    return F_statistic, p_value

# Example: Simulating data for three groups (A, B, C)
np.random.seed(42)  # Set seed for reproducibility
group_A = np.random.normal(10, 2, 30)  # Group A with mean 10, std 2
group_B = np.random.normal(12, 2.5, 30)  # Group B with mean 12, std 2.5
group_C = np.random.normal(15, 3, 30)  # Group C with mean 15, std 3

# Perform the ANOVA test
F_statistic, p_value = anova_test([group_A, group_B, group_C])

# Output the results
print(f"F-statistic: {F_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: At least one group mean is different.")
else:
    print("Fail to reject the null hypothesis: All group means are equal.")
```

#19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

def check_assumptions(groups):
    """
    Check the assumptions for ANOVA: normality, independence, and equal variance.

    Parameters:
    - groups: A list of arrays, each representing a group of data.
    """
    # Assumption 1: Normality
    print("\nAssumption 1: Normality")
    for i, group in enumerate(groups):
        # Shapiro-Wilk test for normality
        stat, p_value = stats.shapiro(group)
        print(f"Group {i+1} - Shapiro-Wilk Test: Stat={stat:.4f}, P-value={p_value:.4f}")

        # If p-value < 0.05, we reject the null hypothesis that the data is normal
        if p_value < 0.05:
            print(f"Group {i+1} does NOT follow a normal distribution (p < 0.05).")
        else:
            print(f"Group {i+1} follows a normal distribution (p >= 0.05).")

        # Q-Q plot to visually check normality
        plt.figure(figsize=(6, 4))
        stats.probplot(group, dist="norm", plot=plt)
        plt.title(f"Q-Q Plot for Group {i+1}")
        plt.show()

    # Assumption 2: Independence
    # The independence assumption is difficult to test directly. However, we can check residuals after ANOVA.
    print("\nAssumption 2: Independence")
    print("Independence is typically ensured by design or study setup.\nVisual inspection of residuals after ANOVA can help verify.")

    # Assumption 3: Equal Variance (Homogeneity of Variances)
    print("\nAssumption 3: Equal Variance")
    # Levene's test for equal variances
    stat, p_value = stats.levene(*groups)
    print(f"Levene's Test: Stat={stat:.4f}, P-value={p_value:.4f}")

    # If p-value < 0.05, we reject the null hypothesis that the variances are equal
    if p_value < 0.05:
        print("The variances are significantly different (p < 0.05). Homogeneity of variances assumption is violated.")
    else:
        print("The variances are equal (p >= 0.05). Homogeneity of variances assumption is met.")

    # Visual check for equal variance (boxplot)
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=groups)
    plt.title("Boxplot of Groups for Visual Check of Equal Variance")
    plt.show()

# Example: Simulating data for three groups (A, B, C)
np.random.seed(42)  # Set seed for reproducibility
group_A = np.random.normal(10, 2, 30)  # Group A with mean 10, std 2
group_B = np.random.normal(12, 2.5, 30)  # Group B with mean 12, std 2.5
group_C = np.random.normal(15, 3, 30)  # Group C with mean 15, std 3

# Check assumptions for ANOVA
check_assumptions([group_A, group_B, group_C])
```

#20.  Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt

def two_way_anova(data, factor1, factor2, dependent_variable):
    """
    Perform a Two-Way ANOVA to study the interaction between two factors.

    Parameters:
    - data: A pandas DataFrame containing the data.
    - factor1: The first factor (independent variable).
    - factor2: The second factor (independent variable).
    - dependent_variable: The dependent variable.

    Returns:
    - ANOVA table: The results of the two-way ANOVA test.
    """
    # Fit the model
    model = ols(f'{dependent_variable} ~ C({factor1}) * C({factor2})', data=data).fit()

    # Perform the two-way ANOVA
    anova_table = sm.stats.anova_lm(model, typ=2)

    return anova_table

# Simulate data for two factors (e.g., Treatment and Age Group)
np.random.seed(42)

# Factor 1: Treatment (2 levels: A and B)
# Factor 2: Age Group (3 levels: Young, Middle, Old)
# Dependent Variable: Test Score

treatment = np.random.choice(['A', 'B'], size=180)  # 180 samples, 2 treatments
age_group = np.random.choice(['Young', 'Middle', 'Old'], size=180)  # 3 age groups
test_score = np.random.normal(50, 10, 180)  # Normally distributed test scores

# Introduce some interaction by adding an offset based on the combination of factors
interaction_effect = np.where((treatment == 'A') & (age_group == 'Young'), 5, 0) + \
                     np.where((treatment == 'B') & (age_group == 'Old'), -5, 0)
test_score += interaction_effect

# Create a DataFrame with the data
data = pd.DataFrame({
    'Treatment': treatment,
    'AgeGroup': age_group,
    'TestScore': test_score
})

# Perform the two-way ANOVA
anova_table = two_way_anova(data, 'Treatment', 'AgeGroup', 'TestScore')

# Output the ANOVA results
print(anova_table)

# Visualization of the interaction effect using a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='AgeGroup', y='TestScore', hue='Treatment', data=data)
plt.title('Interaction Effect of Treatment and Age Group on Test Score')
plt.xlabel('Age Group')
plt.ylabel('Test Score')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def plot_f_distribution(df1, df2):
    """
    Plot the F-distribution for given degrees of freedom (df1, df2).

    Parameters:
    - df1: Degrees of freedom for the numerator (e.g., between-group variability in ANOVA).
    - df2: Degrees of freedom for the denominator (e.g., within-group variability in ANOVA).
    """
    # Generate x values from 0 to 5 with a step of 0.1
    x = np.linspace(0, 5, 1000)

    # Compute the PDF of the F-distribution
    y = stats.f.pdf(x, df1, df2)

    # Plot the F-distribution
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'F-distribution\n(df1={df1}, df2={df2})', color='blue')
    plt.fill_between(x, y, color='skyblue', alpha=0.4)

    # Add labels and title
    plt.title(f'F-distribution with df1={df1} and df2={df2}')
    plt.xlabel('F-statistic')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)

    # Show the plot
    plt.show()

# Example: Visualize the F-distribution for different degrees of freedom
plot_f_distribution(5, 10)  # df1=5, df2=10
plot_f_distribution(10, 20)  # df1=10, df2=20
```

```python
 import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# Simulate data for three different groups
np.random.seed(42)

# Group 1: Treatment A
group_A = np.random.normal(loc=60, scale=10, size=30)

# Group 2: Treatment B
group_B = np.random.normal(loc=65, scale=10, size=30)

# Group 3: Treatment C
group_C = np.random.normal(loc=70, scale=10, size=30)

# Combine the groups into a DataFrame
data = pd.DataFrame({
    'Treatment': ['A']*30 + ['B']*30 + ['C']*30,
    'Score': np.concatenate([group_A, group_B, group_C])
})

# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(group_A, group_B, group_C)

# Output the F-statistic and p-value
print(f'F-statistic: {f_statistic:.4f}')
print(f'p-value: {p_value:.4f}')

# Interpretation of results
if p_value < 0.05:
    print("Reject the null hypothesis: At least one group mean is different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between group means.")

# Visualization of the results with boxplots
plt.figure(figsize=(8, 6))
sns.boxplot(x='Treatment', y='Score', data=data, palette='Set2')
plt.title('Comparison of Treatment Groups using One-Way ANOVA')
plt.xlabel('Treatment Group')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

#23.  Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate random data from a normal distribution
np.random.seed(42)  # For reproducibility

# Simulate data with a mean of 50 and a standard deviation of 10
population_mean = 50
std_dev = 10
sample_size = 100

# Generate random data from a normal distribution
data = np.random.normal(loc=population_mean, scale=std_dev, size=sample_size)

# Step 2: State the hypotheses
# H₀: The population mean is 50 (μ = 50)
# H₁: The population mean is not 50 (μ ≠ 50) - Two-tailed test

# Step 3: Perform a one-sample t-test
t_statistic, p_value = stats.ttest_1samp(data, population_mean)

# Step 4: Output the results
print(f"Sample Mean: {np.mean(data):.2f}")
print(f"Sample Standard Deviation: {np.std(data):.2f}")
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 5: Interpret the result
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the sample mean and the population mean.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")

# Step 6: Visualize the sample data and the population mean
plt.figure(figsize=(8, 6))
plt.hist(data, bins=20, edgecolor='black', alpha=0.7)
plt.axvline(population_mean, color='red', linestyle='dashed', linewidth=2, label=f'Population Mean ({population_mean})')
plt.title('Histogram of Sample Data')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

#24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate random data from a normal distribution
np.random.seed(42)  # For reproducibility

# Assume population variance is 100 (standard deviation = 10)
population_variance = 100
std_dev = np.sqrt(population_variance)
sample_size = 30  # Sample size

# Simulate a sample from a normal distribution
data = np.random.normal(loc=0, scale=std_dev, size=sample_size)

# Step 2: State the hypotheses
# H₀: Population variance is 100 (σ² = 100)
# H₁: Population variance is not 100 (σ² ≠ 100)

# Step 3: Perform Chi-square test for variance
# Calculate the sample variance
sample_variance = np.var(data, ddof=1)  # Sample variance (ddof=1 for unbiased estimator)

# Chi-square test statistic
chi_square_stat = (sample_size - 1) * sample_variance / population_variance

# Degrees of freedom
df = sample_size - 1

# Step 4: Find the critical values for a two-tailed test at 0.05 significance level
alpha = 0.05
critical_value_lower = stats.chi2.ppf(alpha / 2, df)  # Lower critical value
critical_value_upper = stats.chi2.ppf(1 - alpha / 2, df)  # Upper critical value

# Step 5: Output the results
print(f"Sample Variance: {sample_variance:.2f}")
print(f"Chi-square Statistic: {chi_square_stat:.4f}")
print(f"Critical values: Lower = {critical_value_lower:.4f}, Upper = {critical_value_upper:.4f}")

# Interpretation of the results
if chi_square_stat < critical_value_lower or chi_square_stat > critical_value_upper:
    print("Reject the null hypothesis: The population variance is significantly different from 100.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference from 100.")

# Step 6: Visualize the Chi-square distribution and test statistic
x = np.linspace(0, 2 * critical_value_upper, 1000)
y = stats.chi2.pdf(x, df)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'Chi-square distribution\n(df={df})', color='blue')
plt.fill_between(x, y, where=(x > critical_value_upper), color='red', alpha=0.4, label='Rejection Region (Upper)')
plt.fill_between(x, y, where=(x < critical_value_lower), color='red', alpha=0.4, label='Rejection Region (Lower)')
plt.axvline(chi_square_stat, color='green', linestyle='dashed', linewidth=2, label=f'Statistic: {chi_square_stat:.4f}')
plt.title('Chi-square Test for Population Variance')
plt.xlabel('Chi-square Statistic')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

#25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulated Data
# Group 1: 100 successes out of 300 trials
x1 = 100
n1 = 300

# Group 2: 120 successes out of 350 trials
x2 = 120
n2 = 350

# Step 1: Calculate sample proportions
p1 = x1 / n1
p2 = x2 / n2

# Step 2: Calculate the combined proportion
P = (x1 + x2) / (n1 + n2)

# Step 3: Calculate the Z-test statistic
Z = (p1 - p2) / np.sqrt(P * (1 - P) * (1/n1 + 1/n2))

# Step 4: Calculate the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(Z)))  # Two-tailed test

# Step 5: Output the results
print(f"Group 1 Proportion (p1): {p1:.4f}")
print(f"Group 2 Proportion (p2): {p2:.4f}")
print(f"Z-statistic: {Z:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the two proportions.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the two proportions.")

# Step 6: Visualization of the Z-distribution and the critical regions
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Z-distribution', color='blue')
plt.fill_between(x, y, where=(x > 1.96), color='red', alpha=0.4, label='Rejection Region (Right Tail)')
plt.fill_between(x, y, where=(x < -1.96), color='red', alpha=0.4, label='Rejection Region (Left Tail)')
plt.axvline(Z, color='green', linestyle='dashed', linewidth=2, label=f'Z-statistic: {Z:.4f}')
plt.title('Z-Test for Proportions')
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

#26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulate two random datasets (for example: group 1 and group 2)
np.random.seed(42)

# Group 1: Normally distributed data with larger variance
data1 = np.random.normal(loc=0, scale=10, size=100)  # Mean=0, SD=10

# Group 2: Normally distributed data with smaller variance
data2 = np.random.normal(loc=0, scale=5, size=100)   # Mean=0, SD=5

# Step 1: Calculate the sample variances
var1 = np.var(data1, ddof=1)  # Sample variance for group 1
var2 = np.var(data2, ddof=1)  # Sample variance for group 2

# Step 2: Calculate the F-statistic (larger variance / smaller variance)
F_statistic = var1 / var2 if var1 > var2 else var2 / var1

# Step 3: Degrees of freedom
df1 = len(data1) - 1
df2 = len(data2) - 1

# Step 4: Find the critical value for the F-distribution at alpha = 0.05
alpha = 0.05
F_critical = stats.f.ppf(1 - alpha, df1, df2)  # Upper tail critical value

# Step 5: Output the results
print(f"Sample Variance (Group 1): {var1:.4f}")
print(f"Sample Variance (Group 2): {var2:.4f}")
print(f"F-statistic: {F_statistic:.4f}")
print(f"Critical Value: {F_critical:.4f}")

# Interpretation of the result
if F_statistic > F_critical:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")

# Step 6: Visualization of the F-distribution and the F-statistic
x = np.linspace(0, 5, 1000)
y = stats.f.pdf(x, df1, df2)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'F-distribution\n(df1={df1}, df2={df2})', color='blue')
plt.fill_between(x, y, where=(x > F_critical), color='red', alpha=0.4, label='Rejection Region')
plt.axvline(F_statistic, color='green', linestyle='dashed', linewidth=2, label=f'F-statistic: {F_statistic:.4f}')
plt.title('F-Test for Comparing Variances')
plt.xlabel('F-value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

#27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulate observed data (e.g., outcomes of rolling a die 100 times)
np.random.seed(42)
observed_data = np.random.choice([1, 2, 3, 4, 5, 6], size=100, p=[0.1, 0.2, 0.3, 0.1, 0.15, 0.15])

# Step 1: Calculate observed frequencies for each category (die face)
observed_freq = [np.sum(observed_data == i) for i in range(1, 7)]

# Step 2: Define the expected frequencies (assuming a uniform distribution)
# For a fair die, each outcome should occur with equal probability
expected_freq = [100 / 6] * 6  # Since there are 6 possible outcomes, each should have an expected count of 100/6

# Step 3: Perform the Chi-square goodness-of-fit test
chi2_stat, p_value = stats.chisquare(observed_freq, expected_freq)

# Step 4: Output the results
print(f"Observed Frequencies: {observed_freq}")
print(f"Expected Frequencies: {expected_freq}")
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 5: Interpretation of the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data fits the expected distribution.")

# Step 6: Visualization
# Bar plot of observed vs expected frequencies
categories = [1, 2, 3, 4, 5, 6]

plt.figure(figsize=(8, 6))
plt.bar(categories, observed_freq, width=0.4, label='Observed', align='center', alpha=0.7, color='blue')
plt.bar(categories, expected_freq, width=0.2, label='Expected', align='center', alpha=0.7, color='red')
plt.xlabel('Die Faces')
plt.ylabel('Frequency')
plt.title('Chi-square Goodness-of-Fit Test: Observed vs Expected Frequencies')
plt.legend()
plt.grid(True)
plt.show()
```
